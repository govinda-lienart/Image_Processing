# Part 3: Applying Filters

**7.    Blurring (Averaging Filter):**

    Apply a simple 3x3 averaging filter to grayscale_image. For each pixel, replace it with the average of its 3x3 neighborhood.

    Repeat for rgb_image by applying the filter separately to each channel (R, G, B).

```python
from scipy.ndimage import convolve
```

```python
kernel = np.ones((3, 3)) / 9
blurred_grayscale = convolve(rotated_grayscale, kernel)
# Create a 3x3 averaging filter (or kernel) to apply a blurring effect
# np.ones((3, 3)) creates a 3x3 matrix filled with 1s:
# [[1, 1, 1],
#  [1, 1, 1],
#  [1, 1, 1]]
# After division, the kernel becomes:
# [[1/9, 1/9, 1/9],
#  [1/9, 1/9, 1/9],
#  [1/9, 1/9, 1/9]]
# This means each pixel in the 3x3 neighborhood contributes equally to the average.
# When this kernel is applied to the image, it will calculate the average of
# each 3x3 region, replacing the center pixel with the average, which produces a blur effect.
```

```python
blurred_rgb = np.zeros_like(rotated_rgb)


 #this line initialize an empty RGB image with the same shape as the original RGB image.
# This empty image (blurred_rgb) is then used to store the blurred result for each color channel.
# np.zeros_like(rgb_image) creates an array with the same dimensions as rgb_image, but filled with zeros.
# For an RGB image, this means we have three channels (R, G, B), each set to zero initially.

for i in range(3):
      blurred_rgb[..., i] = convolve(rotated_rgb[..., i], kernel)

# Loop over each channel in the RGB image.
# Here, 'i' represents the channel index:
# i=0 for the Red channel
# i=1 for the Green channel
# i=2 for the Blue channel

    # so to this will provide Access the current channel (either R, G, or B) of the original RGB image.
    # rgb_image[..., i] selects all rows and columns (indicated by ...), but only for channel 'i'.
    # This results into a 2D array for the selected color channel.

    # Apply the 3x3 averaging filter to this 2D channel using the convolve function.
    # convolve(rgb_image[..., i], kernel) computes the average for each 3x3 neighborhood
    # and replaces the center pixel with this average value, resulting in a blurred effect.

    # Store the blurred result for this channel in the corresponding channel of blurred_rgb.
    # blurred_rgb[..., i] will hold the blurred values for channel 'i' (R, G, or B), replacing zeros.

# After the loop finishes, blurred_rgb contains the blurred results for each color channel (R, G, B).
# Now, blurred_rgb is a complete RGB image, with each channel blurred individually.
# This array eventually displayed or saved as a combined, fully blurred RGB image.
```

```python

# Display the images side by side
plt.figure(figsize=(12, 6))

# Display the original grayscale image
plt.subplot(2, 2, 1)
plt.imshow(grayscale_image, cmap='gray')
plt.title("Original Grayscale Image")

# Display the blurred grayscale image
plt.subplot(2, 2, 2)
plt.imshow(blurred_grayscale, cmap='gray')
plt.title("Blurred Grayscale Image")

# Display the original RGB image
plt.subplot(2, 2, 3)
plt.imshow(rotated_rgb)
plt.title("Original RGB Image")

# Display the blurred RGB image
plt.subplot(2, 2, 4)
plt.imshow(blurred_rgb)
plt.title("Blurred RGB Image")
# Show all images
plt.tight_layout()
plt.show()

plt.show()
```

# 8.    Edge Detection (Simple Gradient Filter):

    Create a gradient filter (e.g., using a simple horizontal edge detection kernel like [-1, 0, 1]) and apply it to grayscale_image.


```python
# Define the horizontal edge detection kernel
horizontal_kernel = np.array([[-1, 0, 1]])

# Apply the horizontal edge detection kernel to the blurred grayscale image. np.abs(convolve(...)) enhances
# edge detection by applying a filter to detect changes, then converting all values to positive with np.abs.
# This makes edges equally visible, regardless of direction, highlighting boundaries and structures clearly in the image.
edges_blurred_grayscale = np.abs(convolve(blurred_grayscale, horizontal_kernel))

# Apply the horizontal edge detection kernel separately to each channel of the blurred RGB image
edges_blurred_rgb = np.zeros_like(blurred_rgb)
for i in range(3):  # Loop over each channel (R, G, B)
    edges_blurred_rgb[..., i] = np.abs(convolve(blurred_rgb[..., i], horizontal_kernel))

# Display the results
plt.figure(figsize=(12, 6))

# Display the original blurred grayscale image
plt.subplot(2, 2, 1)
plt.imshow(blurred_grayscale, cmap='gray')
plt.title("Blurred Grayscale Image")

# Display the enhanced horizontal edges in the blurred grayscale image
plt.subplot(2, 2, 2)
plt.imshow(edges_blurred_grayscale_enhanced, cmap='gray')
plt.title("Enhanced Horizontal Edges (Blurred Grayscale)")

# Display the original blurred RGB image
plt.subplot(2, 2, 3)
plt.imshow(blurred_rgb)
plt.title("Blurred RGB Image")

# Display the horizontal edges in the blurred RGB image (kept in grayscale)
plt.subplot(2, 2, 4)
plt.imshow(edges_blurred_rgb)
plt.title("Horizontal Edges (Blurred RGB)")

plt.tight_layout()
plt.show()
```
